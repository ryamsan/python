```python
#from matplotlib import pyplot as plt 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph
plt.figure(figsize=(5,3), dpi=150)

## Line 1

# plt.plot(x,y, label='2x', color='red', linewidth=3, marker='.', markersize=12, markeredgecolor='blue', linestyle='--' ) # or #ababab or y
# OR plt.plot(x,y)

# Use shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2

#select interval we want to plot the points at 
x2 = np.arange(0,4.5,0.5)
print(x2)

# Plot part of the graph as line
plt.plot(x2[:5], x2[:5]**2, 'r.-', label='X^2')

# Plot remainder of graph
plt.plot(x2[4:], x2[4:]**2, 'g--', label='X^2')


plt.title('Our first graph!', fontdict={'fontname': 'Comic Sans Ms', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

#change the sizing of the ticks
plt.xticks([0,1,2,3,4])
plt.yticks(range(0,17,2))

plt.legend()
# OR plt.legend('2x')

plt.savefig('mygraph.png', dpi=300)

plt.show()
```

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
# OR    
#bars[0].set_hatch('/')
#bars[1].set_hatch('O')
#bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```
